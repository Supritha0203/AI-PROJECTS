```python
import cv2
faceCascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
image=cv2.imread("fd pic.jpg")
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(image)
```

```python
rgb_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
gr_img=cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gr_img)
```

```python
faces=faceCascade.detectMultiScale(gr_img,scaleFactor = 1.2,minNeighbors = 5,minSize = (30,30))
print(faces)
```

```python
image=cv2.imread("fd pic.jpg")
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),20)
rgb_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```
