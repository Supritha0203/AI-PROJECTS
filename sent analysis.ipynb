```python
def preprocess_text(text):
  text= text.lower()
  text= re.sub("@[\w*","",text)
  text= re.sub("http\S+","",text)
  text= re.sub("[^a-zA-Z#","",text)
  text= re.sub("rt","",text)
  text= re.sub("\s+"," ",text)

  return text
```
